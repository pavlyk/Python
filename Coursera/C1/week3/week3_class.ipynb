```
class Planet:
    def __new__(cls, *args, **kwargs):
        print("__new__ called")
        obj = super().__new__(cls)
        return obj
    def __init__(self, name):
        print("__init__ called")
        self.name = name
earth = Planet("Earth")
```

В этом примере в исходный метод __new__ был добавлен вызов print(). В следующей строке метод super() возвращает родителя нашего класса, в данном случае это
object --- класс, от которого наследуются все пользовательские классы в Python 3. Затем
вызывается метод __new__ класса object, который возвращает экземляр класса. Этот
экземпляр (который и является нашим классом) возвращается из функции.
